### Soru 1: Sayının İşareti
Girilen bir sayının pozitif, negatif veya sıfır olduğunu bulan programı yazınız.

```python
# Kodunuzu buraya yazınız

sayı =int(input("lütfen bir sayı gitiniz...")) 
if sayı>0:
    print("Pozitif")
elif sayı<0:
    print("negatif")
else : 
    print("sayı sıfır")
```

### Soru 3: Sesli Harf Kontrolü
Girilen bir harfin sesli harf (a,e,i,o,u) olup olmadığını kontrol eden programı yazınız.

```python
# Kodunuzu buraya yazınız
girilenHarf=input(str("lütfen bir harf giriniz"))

if girilenHarf=="a":
    print("girilen a harfi ünlü bir harftir")
elif girilenHarf=="e":
     print("girilen e harfi ünlü bir harftir")
elif girilenHarf=="i":
     print("girilen i harfi ünlü bir harftir")        
elif girilenHarf=="o":
     print("girilen u harfi ünlü bir harftir")
elif girilenHarf=="u":
     print("girilen u harfi ünlü bir harftir")
else:
     print("girilen harf ünlü bir harf değildir")
          
```

### Soru 6: Not Harf Karşılığı

### Soru 7: Artık Yıl Hesaplama
Girilen bir yılın artık yıl olup olmadığını hesaplayan programı yazınız.
(Artık yıl: 4'e tam bölünen, ancak 100'e tam bölünüyorsa 400'e de tam bölünmeli)

```python
# Kodunuzu buraya yazınız
girilenyıl=int(input("lütfen bir yıl giriniz...."))

if girilenyıl%100==0 and girilenyıl%400==0:
    print("artık yıldır")
elif girilenyıl%4==0:
    print("girilen yıl artıktır")     
else:
    print("girirlen yıl artık yıl değildir...")       
```

### Soru 8: Karakter Türü Belirleme
Girilen bir karakterin rakam, harf veya özel karakter olduğunu bulan programı yazınız.

```python
# Kodunuzu buraya yazınız
girilenKarakter=input("lütfen bir karakter giriniz...")

if girilenKarakter.isdigit():
    print("girilen karakter rakam")
elif girilenKarakter.isalpha():
    print("girilen karakter harf")
else:
     print("girilen karakter özel karakter")

```
